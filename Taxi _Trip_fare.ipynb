# Calculation of trip distance

```python
import math
from math import radians, sin, cos, asin, sqrt
```

```python
#Using Haversine distance formula to calculate trip distance

def haversine_distance(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance between two points on Earth (in miles)
    using the Haversine formula
    """
    # Convert decimal degrees to radians
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])

    # Haversine formula
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2) ** 2 + cos(lat1) * cos(lat2) * sin(dlon / 2) ** 2
    c = 2 * asin(sqrt(a))

    # Radius of Earth in miles
    r = 3963
    return c * r
```

```python
#pickup time
df['pickup_year'] = df['tpep_pickup_datetime'].dt.year
df['pickup_month'] = df['tpep_pickup_datetime'].dt.month
df['pickup_weekday_name'] = df['tpep_pickup_datetime'].dt.day_name()
df['pickup_hour'] = df['tpep_pickup_datetime'].dt.hour

# Dropoff time
df['dropoff_year'] = df['tpep_dropoff_datetime'].dt.year
df['dropoff_month'] = df['tpep_dropoff_datetime'].dt.month
df['dropoff_weekday_name'] = df['tpep_dropoff_datetime'].dt.day_name()
df['dropoff_hour'] = df['tpep_dropoff_datetime'].dt.hour


# Feature engineering function
def feature_engineering(df):
    """
    Create new features from existing columns
    """
    print("🔧 Starting Feature Engineering...")
    data_clean = df.copy()

    # Calculate trip distance using Haversine formula
    print("📏 Calculating trip distances...")
    data_clean['trip_distance_calculated'] = data_clean.apply(
        lambda row: haversine_distance(
            row['pickup_longitude'], row['pickup_latitude'],
            row['dropoff_longitude'], row['dropoff_latitude']
        ), axis=1
    )

    # Include weekday names from the original DataFrame
    data_clean['pickup_weekday_name'] = df['pickup_weekday_name']
    data_clean['dropoff_weekday_name'] = df['dropoff_weekday_name']


    print("✅ Feature engineering completed!")
    print(f"📊 New dataset shape: {data_clean.shape}")

    return data_clean

# Example usage:
# df_engineered will have all new features
df_engineered = feature_engineering(df)

# Now use df_engineered instead of data_clean
df = df_engineered.copy()

print(df.columns)
```

## 5️⃣ Data Transformation:

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer

# ===============================
# STEP 5 : Data Transformation
# ===============================

print("🔧 Starting Data Transformation...")

# 1. Remove invalid rows
before_rows = df.shape[0]

df = df[
    (df['total_amount'] > 0) &
    (df['trip_distance_calculated'] > 0) &
    (df['passenger_count'] > 0)
]

after_rows = df.shape[0]
print(f"🧹 Removed {before_rows - after_rows} invalid rows.")

# 2. Handle Outliers (using IQR method for fare & distance)

def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[column] >= lower) & (df[column] <= upper)]

df = remove_outliers_iqr(df, 'total_amount')
df = remove_outliers_iqr(df, 'trip_distance_calculated')

print(f"✅ After outlier removal: {df.shape}")

# 3. Select Features for ML

categorical_features = ['payment_type', 'RatecodeID', 'is_pm', 'is_night',
                        'is_morning_rush', 'is_evening_rush']
numerical_features = ['trip_distance_calculated','trip_duration_calculated',
                      'passenger_count','pickup_hour','dropoff_hour',
                      'pickup_month','dropoff_month']

X = df[categorical_features + numerical_features]
y = df['total_amount']

# 4. Preprocessing Pipeline (Encoding + Scaling)

categorical_transformer = OneHotEncoder(handle_unknown='ignore')
numerical_transformer = StandardScaler()

preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, categorical_features),
        ('num', numerical_transformer, numerical_features)
    ])

print("✅ Data Transformation Completed!")
```
